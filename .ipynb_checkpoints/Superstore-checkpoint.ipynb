<strong id="#Importinng Libraries">Importinng Libraries<strong/>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import date
```

<strong id="#Load Data">Load Data<strong/>

```python
df = pd.read_csv(r'C:\laragon\www\super_store\superstore_sales.csv', encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

<strong id="#Data Cleaning">Data Cleaning<strong/>

```python
df.drop_duplicates()
```

```python
df.isnull().sum()
```

<strong id="#Change Data Types">Change Data Types<strong/>

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df['sales'] = pd.to_numeric(df['sales'], errors='coerce').fillna(0, downcast='infer')
```

```python
df.dtypes
```

```python
df.describe()
```

<strong id="#Exploratory Data Analyst">Exploratory Data Analyst<strong/>

```python
ax = sns.countplot(x='ship_mode', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby('ship_mode')['quantity'].sum().sort_values(ascending=False)
```

```python
sales_ship_mode = df.groupby(['ship_mode'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)

sns.barplot(x='ship_mode', y='quantity', data=sales_ship_mode)
```

```python
df.groupby('ship_mode')['profit'].count().sort_values(ascending=False)
```

```python
sns.barplot(y='ship_mode', x='profit', data=df)
plt.xticks(rotation=5)
plt.grid(True)
```

```python
df.sort_values(by='quantity', ascending=False)['customer_name'].head(10)
```

```python
name_top10 = df.sort_values(by='quantity', ascending=False).head(10)
```

```python
plt.figure(figsize = (20, 10))
sns.barplot(y='quantity', x='customer_name', data=name_top10)
plt.xticks(rotation=3)
plt.grid(True)
```

```python
df.groupby('customer_name')['profit'].sum().sort_values(ascending=False)
```

```python
df.sort_values(by='profit', ascending=False)['customer_name'].head(10)
```

```python
top10 = df.sort_values(by='profit', ascending=False).head(10)
```

```python
sns.stripplot(y="customer_name", x="profit", data=top10)
plt.grid(True)
```

```python
ax = sns.countplot(x='segment', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_segment = df.groupby(['segment'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)
3q1
sns.barplot(x='segment', y='quantity', data=sales_segment)
```

```python
top10 = df.sort_values(by='country', ascending=False).head(10)
```

```python
ax = sns.countplot(x='state', data=top10)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.barplot(y='profit', x='state', data=top10)
plt.xticks(rotation=5)
plt.grid(True)
```

```python
ax = sns.countplot(x='market', data=df)

for plot in ax.containers:
    ax.bar_label(plot)
```

```python
sales_market = df.groupby(['market'],as_index=False)['quantity'].sum().sort_values(by='quantity', ascending=False)

sns.barplot(x='market', y='quantity', data=sales_market)
```
